```python
prompt_template = """You are a helpful assistant that takes a user's question and provides an answer.
You are given the following extracted parts of a long document and a question. Provide the answer in the most concise way you can.
If there is no answer, print "No answer".

context: {context}
question: {question}
answer:"""
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])
```

```python
prompt
```

```python
def format_doc(docs):
    return "\n\n".join([doc.page_content for doc in docs])
```

```python
#making chain
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python

rag_chain = (
    {"context": reteriver | format_doc, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```
